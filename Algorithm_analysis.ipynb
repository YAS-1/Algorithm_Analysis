```python
import matplotlib.pyplot as plt
import time
import numpy as np

def swap(lst, i, j):
    """Swaps elements at indices i and j in list lst"""
    temp = lst[i]
    lst[i] = lst[j]
    lst[j] = temp
    return lst

#INSERTION SORT
def insertion_sort(arr):
    """Sorts an array arr using insertion sort"""
    n = len(arr)
    for i in range(1, n):
        for j in range(i, 0, -1):
            if arr[j - 1] > arr[j]:
                swap(arr, j - 1, j)
    return arr

#MERGE SORT
def merge_sort(arr):
    """Sorts an array arr using merge sort"""
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left_half = merge_sort(arr[:mid])
    right_half = merge_sort(arr[mid:])

    return merge(left_half, right_half)

def merge(left, right):
    """Merges 
 two sorted lists left and right"""
    result = []
    i1 = i2 = 0
    while i1 < len(left) and i2 < len(right):
        if left[i1] <= right[i2]:
            result.append(left[i1])
            i1 += 1
        else:
            result.append(right[i2])
            i2 += 1
    result.extend(left[i1:])
    result.extend(right[i2:])
    return result

#QUICK SORT
def quick_sort(arr):
    """Sorts an array arr using quick sort"""
    if len(arr) <= 1:
        return arr

    n = len(arr)
    pivot = arr[n // 2]
    L = [x for x in arr if x < pivot]
    R = [x for x in arr if x > pivot]
    return quick_sort(L) + [pivot] + quick_sort(R)

#SELECTION SORT
def selection_sort(array):
    """Sorts an array array using selection sort"""
    for i in range(0,len(array)-1):
        minposition = i

        for j in range(i+1,len(array)):
            if array[j] < array[minposition]:
                minposition = j
        
        if minposition != i:
            array[i], array[minposition] = array[minposition], array[i]
    return array

#BUBBLE SORT
def bubble_sort(array):
    """Sorts an array array using bubble sort"""
    for i in range(len(array)-1, 0, -1):
        for j in range(i):
            if array[j] > array[j+1]:
                array[j], array[j+1] = array[j+1], array[j]

    return array

sorts = [
    {
        "name": "Insertion Sort",
        "sort": insertion_sort
    },
    {
        "name": "Merge Sort",
        "sort": merge_sort
    },
    {
        "name": "Quick Sort",
        "sort": quick_sort
    },
    {
        "name": "Selection Sort",
        "sort": selection_sort
    },
    {
        "name": "Bubble Sort",
        "sort": bubble_sort
    }
]

elements = np.array([i * 1000 for i in range(1, 15)])

plt.xlabel('Size of n')
plt.ylabel('Time Complexity')

for sort in sorts:
    times = list() #times = list(): Initializes an empty list to store the time taken by the current sorting algorithm for different input sizes.
    start_all = time.time() #start_all = time.time(): Records the start time for the entire sorting experiment

    # Inner Loop (for i in range(1, 15)): This loop runs for each input size (from 1000 to 14000 in increments of 1000).
    for i in range(1, 15):
        start = time.time()
        a = np.random.randint(1000, size=i * 1000) #Generates a random array of integers with size i * 1000.
        sort["sort"](a)
        end = time.time()
        times.append(end - start)# Calculates and stores the time taken to sort the array by subtracting start from end.
    end_all = time.time()

    plt.plot(elements, times, label=sort["name"])

plt.grid()
plt.legend()
plt.show()
```
